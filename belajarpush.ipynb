1. `git clone` + [ssh link](git@github.com:KevinDHS/KevinDHS.git)
2. buat **file** baru di folder tsb / pindahkan **file** yg mau di push
3. make sure file nya sudah pindah di folder ssh github yg di clone 
4. `git status` -> harusnya muncul nama file dan berwarna merah
5. `git add .` -> memasukkan semua file ke dalam main branch
6. `git status ` -> cek lagi file nya, seharusnya menjadi hijau 
7. `git commit -m "final"` -> -m itu adalah message, yang dalam tanda kutip artinya keterangan
8. `git push`
